# Simple example of the anomaly detection on the synthetic data with outliers

## Libraries importing

```python
import pandas as pd
import numpy as np

import sys
sys.path.append('../')
from ControlCharts import T2
```

## T2 without PCA

### Data generating

```python
df = pd.DataFrame(np.random.randn(100, 4), columns=list('ABCD'))
```

```python
df.plot(figsize=(14, 3))
```

### Adding outliers

```python
df.at[80, 'A'] = 12
df.at[46, 'B'] = -6.5
df.at[93, 'C'] = 4.5
df.at[30, 'D'] = -20
```

```python
df.plot(figsize=(14, 3))
```

### Method initialization

```python
t2 = T2(using_pca=False)
```

### Method fitting

```python
t2.fit(df.iloc[:20])
```

### Method applying

```python
t2.predict(df)
```

3 out of 4 outliers were detected
